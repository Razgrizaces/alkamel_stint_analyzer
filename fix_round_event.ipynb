```python
import pandas as pd
```

```python
events = pd.read_csv('events_mapping.csv')
```

```python
x_hour_events = events[(events['championship']=="ELMS")|(events['championship']=="FIAWEC")]
x_hour_events['event'] = x_hour_events['elapsed_hour'].map(str) + " hours of " + x_hour_events['circuit']
x_hour_events['event_key'] = x_hour_events['championship'] + "_" + x_hour_events['season'] + "_" + x_hour_events['round'].map(str)
x_hour_events.to_csv('x_hour_events.csv')
```
