```python
import numpy as np
import pandas as pd
new_data = pd.read_csv('alkamel_data_timing_board.csv')
new_data
```

```python
new_data_2023 = new_data[new_data['season']==2023]
#new_data_2023 = new_data
new_data_2023 = new_data_2023.sort_values(by=['season','championship','round','number'])
#since we're mostly working with just race data probably better to just go wtih using season + round + number
new_data_2023['team_key'] = new_data_2023['season'].map(str) + "_" + new_data_2023['championship'] + "_" + new_data_2023['round'].map(int).map(str) + "_" + new_data_2023['number'].map(str)
new_data_2023['driver_key'] = new_data_2023['season'].map(str) + "_" + new_data_2023['championship'] + "_" + new_data_2023['round'].map(int).map(str) + "_" + new_data_2023['number'].map(str) + "_" + new_data_2023['driver_number'].map(str)
new_data_2023 = new_data_2023.reset_index(drop=True)
new_data_2023
```

```python
def add_team_stint(new_data):
    #we take the next lap: next = df shifted by 1 row (starts from lap 2)
    new_data_next = new_data.shift(-1).fillna(0)

    #check where the team changes
    new_data['change_team'] = np.where(new_data['number'] == new_data_next['number'], 0, 1)
    new_data['change_team'] = new_data['change_team'].shift(1).fillna(0)

    new_data['change_stint'] = np.where(new_data['crossing_finish_line_in_pit'] == "B", 1, 0)
    new_data['change_stint'] = new_data['change_stint'].shift(1).fillna(0)
        
    #here we're creating a dictionary for the keys by pivoting it by these columns
    wanted_columns = ['team_key']
    new_data_pivoted = new_data.groupby(wanted_columns).mean().reset_index()[wanted_columns]
    new_data_pivoted['team_stint_no'] = 0

    wanted_columns = ['team_key', 'team_stint_no']
    new_data_pivoted_wc = new_data_pivoted[wanted_columns]
    new_data_pivoted_dict = new_data_pivoted_wc.to_dict('records')

    #methods for incrementing and getting values of keys in dictionaries
    def search_key_in_dicts(key, dict):
        for d in dict:
            if d['team_key'] == key:
                return d
        return None
    def update_value_in_dicts(key, dict, col, value):
        dict_key = search_key_in_dicts(key, dict)
        dict_key.update({col : value})
    def increment_driver_stint_value(key, dict):
        update_value_in_dicts(key, dict, 'driver_stint_no', search_key_in_dicts(key, dict).get('driver_stint_no') + 1)
        return search_key_in_dicts(key, new_data_pivoted_dict).get('driver_stint_no')
    def increment_team_stint_value(key,dict):
        update_value_in_dicts(key, dict, 'team_stint_no', search_key_in_dicts(key, dict).get('team_stint_no') + 1)
        return search_key_in_dicts(key, new_data_pivoted_dict).get('team_stint_no')

    key = new_data.at[0, 'team_key']
    #zero everything out again, but we need to increment the very first stint
    update_value_in_dicts(key, new_data_pivoted_dict, 'team_stint_no', 1)
    #getting team_stint_no
    new_data['team_stint_no'] = 0
    for i in range(0,len(new_data)):
        x = new_data.iloc[i]
        if x['change_stint'] == 1 or x['change_team'] == 1:
            increment_team_stint_value(x.team_key, new_data_pivoted_dict)
        new_data.at[x.name, 'team_stint_no'] = (search_key_in_dicts(x.team_key, new_data_pivoted_dict).get('team_stint_no'))

    #so using the same key parameter we zero it all out again
    wanted_columns = ['driver_key']
    new_data_pivoted = new_data.groupby(wanted_columns).mean().reset_index()[wanted_columns]
    new_data_pivoted['driver_stint_no'] = 0

    #and then convert these into a dictionary
    wanted_columns = ['driver_key', 'driver_stint_no']
    new_data_pivoted_wc = new_data_pivoted[wanted_columns]
    new_data_pivoted_dict = new_data_pivoted_wc.to_dict('records')

    #zero everything out again, but we need to increment the very first stint
    key = new_data.at[0, 'driver_key']
    update_value_in_dicts(key, new_data_pivoted_dict, 'driver_stint_no', 1)

    new_data['driver_stint_no'] = 0
    for i in range(0,len(new_data)):
        x = new_data.iloc[i]
        if x['change_stint'] == 1:
            increment_driver_stint_value(x.driver_key, new_data_pivoted_dict)
        new_data.at[x.name, 'driver_stint_no'] = (search_key_in_dicts(x.driver_key, new_data_pivoted_dict).get('driver_stint_no'))
    return new_data

new_data_2023_w_team_stint = add_team_stint(new_data_2023)
new_data_2023_w_team_stint
```

```python
new_data_2023_w_team_stint['team_stint'] = new_data_2023_w_team_stint['team_no'] + " Stint #" + new_data_2023_w_team_stint['team_stint_no'].map(str)
new_data_2023_w_team_stint['driver_stint'] = new_data_2023_w_team_stint['driver_name'] + " Stint #" + new_data_2023_w_team_stint['team_stint_no'].map(str)

new_data_2023_w_team_stint.to_csv('2023_data_with_stint_no.csv')
```
