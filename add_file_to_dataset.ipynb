```python
import numpy as np
import pandas as pd
```

```python
#this is to help me create a new pipeline for the fia_wec dataset
#also shows how the data gets put into something usable.

#this is currently the data that we're taking a look at, we're only really curious about 2022 since that's the lastest year.
fia_wec_data = pd.read_csv("2012-2022_FIA_WEC_FULL_LAP_DATA.csv", index_col=0)
#have to map the season to str bc of the way it displays. 
fia_wec_data.season = fia_wec_data.season.map(str)
fia_wec_data_2022 = fia_wec_data[fia_wec_data['season'] == "2022"].reset_index(drop=True)
fia_wec_data_2022
```

```python
#reading the new file in, this time it's lemans.
new_data = pd.read_csv('2022_LEMANS.csv', sep = ';')
new_data = new_data.dropna(axis=1, how="all")
#have to strip + and convert the columns in the data
new_data.columns = new_data.columns.str.lower().str.strip()
new_data.columns
new_data
```

```python
#adding season, circuit, round
new_data['season'] = '2022'
new_data['circuit'] = 'LE MANS'
new_data['round'] = 3
```

```python
#checks what columns we need to add
full_data_columns = pd.DataFrame(fia_wec_data_2022.columns)
full_data_columns['in_new_df'] = fia_wec_data_2022.columns.isin(new_data.columns)
full_data_columns[full_data_columns['in_new_df'] == False]
```

```python
#this contains the mapping for the vehicles
fia_wec_engine_mapping = pd.read_csv("fia_wec_engine_mapping.csv", index_col = 0)
#only getting the engine/vec map for 2022
fia_wec_veh_to_eng_2022 = fia_wec_engine_mapping[fia_wec_engine_mapping['season'] == "2022"].reset_index(drop=True)
fia_wec_veh_to_eng_2022 = fia_wec_veh_to_eng_2022.groupby(['manufacturer', 'class', 'vehicle', 'engine']).mean()
fia_wec_veh_to_eng_2022 = fia_wec_veh_to_eng_2022.reset_index().drop('number', axis=1)
```

```python
#showing what we're merging on, manufacturer + class
new_data.groupby(['manufacturer', 'class']).mean().reset_index()
```

```python
#join the two based on the manufacturer/class, left join in case there are any missing/fallouts
new_data = pd.merge(new_data, fia_wec_veh_to_eng_2022, on = ['manufacturer', 'class'], how = 'left')
new_data
```

```python
#checks what columns we still need to add
full_data_columns = pd.DataFrame(fia_wec_data_2022.columns)
full_data_columns['in_new_df'] = fia_wec_data_2022.columns.isin(new_data.columns)
full_data_columns[full_data_columns['in_new_df'] == False]
```

```python
#we need: 'team_no', 'lap_time_ms', 'lap_time_s', 'driver_stint_no', 'driver_stint', 'team_stint_no', 'team_stint'
def convert_to_ms(x):
    hour = 0
    try:
        hour, minutes, second = x.split(":")
    except ValueError:
        minutes, second = x.split(":")
    hour_seconds = int(hour)*60*60
    seconds, ms = second.split(".")
    minute_seconds = (int(minutes)*60 + int(seconds) + int(hour_seconds))*1000 
    total_ms = minute_seconds + int(ms)
    return total_ms

#team_no
new_data['team_no'] = new_data['team'] + " #" + new_data['number'].map(int).map(str)

#lap_time_ms + lap_time_s
#we can then apply this function to every single lap time in the df, and convert everything to ms.
new_data['lap_time_ms'] = new_data['lap_time'].apply(convert_to_ms)
#and then let's make a seconds column to plot it neatly later.
new_data['lap_time_s'] = new_data['lap_time_ms']/1000
new_data
```

```python
#getting team and driver stint nos

#we have two different keys for driver and team, this one is for team
new_data['key'] = new_data['season'].map(int).map(str) + "_" + new_data['round'].map(int).map(str) + "_" + new_data['number'].map(int).map(str)

#we take the next lap: next = df shifted by 1 row (starts from lap 2)
new_data_next = new_data.shift(-1).fillna(0)

#check where the driver changes, different from "crossing line in pit" since we know that the driver changes.
new_data['change_driver'] = np.where(new_data['key'] == new_data_next['key'], 0, 1)
new_data['change_driver'] = new_data['change_driver'].shift(1).fillna(0)

#check where the team changes
new_data['change_team'] = np.where(new_data['number'] == new_data_next['number'], 0, 1)
new_data['change_team'] = new_data['change_team'].shift(1).fillna(0)

new_data['change_stint'] = np.where(new_data['crossing_finish_line_in_pit'] == "B", 1, 0)
new_data['change_stint'] = new_data['change_stint'].shift(1).fillna(0)

#here we're creating a dictionary for the keys by pivoting it by these columns
wanted_columns = ['key']
new_data_pivoted = new_data.groupby(wanted_columns).mean().reset_index()[wanted_columns]
new_data_pivoted['team_stint_no'] = 0

wanted_columns = ['key', 'team_stint_no']
new_data_pivoted_wc = new_data_pivoted[wanted_columns]
new_data_pivoted_dict = new_data_pivoted_wc.to_dict('records')

#methods for incrementing and getting values of keys in dictionaries
def search_key_in_dicts(key, dict):
    for d in dict:
        if d['key'] == key:
            return d
    return None
def update_value_in_dicts(key, dict, col, value):
    dict_key = search_key_in_dicts(key, dict)
    dict_key.update({col : value})
def increment_driver_stint_value(key, dict):
    update_value_in_dicts(key, dict, 'driver_stint_no', search_key_in_dicts(key, dict).get('driver_stint_no') + 1)
    return search_key_in_dicts(key, new_data_pivoted_dict).get('driver_stint_no')
def increment_team_stint_value(key,dict):
    update_value_in_dicts(key, dict, 'team_stint_no', search_key_in_dicts(key, dict).get('team_stint_no') + 1)
    return search_key_in_dicts(key, new_data_pivoted_dict).get('team_stint_no')

key = new_data.at[0, 'key']
#zero everything out again, but we need to increment the very first stint
update_value_in_dicts(key, new_data_pivoted_dict, 'team_stint_no', 1)
#getting team_stint_no
new_data['team_stint_no'] = 0
for i in range(0,len(new_data)):
    x = new_data.iloc[i]
    if x['change_stint'] == 1 or x['change_team'] == 1:
        increment_team_stint_value(x.key, new_data_pivoted_dict)
    new_data.at[x.name, 'team_stint_no'] = (search_key_in_dicts(x.key, new_data_pivoted_dict).get('team_stint_no'))
```

```python
#we need a new key for driver number and driver stint
new_data['key'] = new_data['season'].map(int).map(str) + "_" + new_data['round'].map(int).map(str) + "_" + new_data['number'].map(int).map(str) + "_"+ new_data['driver_number'].map(int).map(str)

#so using the same key parameter we zero it all out again
wanted_columns = ['key']
new_data_pivoted = new_data.groupby(wanted_columns).mean().reset_index()[wanted_columns]
new_data_pivoted['driver_stint_no'] = 0

#and then convert these into a dictionary
wanted_columns = ['key', 'driver_stint_no']
new_data_pivoted_wc = new_data_pivoted[wanted_columns]
new_data_pivoted_dict = new_data_pivoted_wc.to_dict('records')

#zero everything out again, but we need to increment the very first stint
key = new_data.at[0, 'key']
update_value_in_dicts(key, new_data_pivoted_dict, 'driver_stint_no', 1)

new_data['driver_stint_no'] = 0
for i in range(0,len(new_data)):
    x = new_data.iloc[i]
    if x['change_stint'] == 1:
        increment_driver_stint_value(x.key, new_data_pivoted_dict)
    new_data.at[x.name, 'driver_stint_no'] = (search_key_in_dicts(x.key, new_data_pivoted_dict).get('driver_stint_no'))
```

```python
new_data['driver_stint'] = new_data['driver_name'] + " Stint #" +  new_data['driver_stint_no'].map(int).map(str)
new_data['team_stint'] = new_data['team_no'] + " Stint #" +  new_data['team_stint_no'].map(int).map(str)
#dropping s1_seconds, since the fia wec data seems to add new columns every iteration ;;
new_data = new_data.drop(['change_driver', 'change_team', 'change_stint', 'key', 's1_seconds', 's2_seconds', 's3_seconds'], axis=1)
```

```python
#checking that team stint + driver stint are right
new_data
```

```python
#hour isn't right, so we have to make elapsed/elapsedms
def convert_to_ms(x):
    hour = 0
    try:
        hour, minutes, second = x.split(":")
    except ValueError:
        minutes, second = x.split(":")
    hour_seconds = int(hour)*60*60
    seconds, ms = second.split(".")
    minute_seconds = (int(minutes)*60 + int(seconds) + int(hour_seconds))*1000 
    total_ms = minute_seconds + int(ms)
    return total_ms
new_data['elapsed_ms'] = new_data.elapsed.map(convert_to_ms)
new_data['elapsed_s'] = new_data.elapsed.map(convert_to_ms)/1000
new_data_with_filters_sorted_hour = new_data.sort_values(['round', 'lap_number', 'elapsed_s']).reset_index(drop=True)
new_data_with_filters_sorted_hour
```

```python
#gets dict for sorted hour by overall
new_data_with_filters_sorted_hour['key'] = new_data_with_filters_sorted_hour['season'] + "_" +  new_data_with_filters_sorted_hour['round'].map(int).map(str) + "_" + new_data_with_filters_sorted_hour['lap_number'].map(int).map(str) 
wanted_columns = ['key', 'round', 'lap_number']
data_keys = new_data_with_filters_sorted_hour.groupby('key').mean().reset_index()[wanted_columns].sort_values(['round','lap_number']).reset_index(drop=True)
data_keys['position'] = 0
data_keys_dict = data_keys.to_dict('records')

def search_key_in_dicts(key, dict):
    for d in dict:
        if d['key'] == key:
            return d
    return None
def update_value_in_dicts(key, dict, col, value):
    dict_key = search_key_in_dicts(key, dict)
    dict_key.update({col : value})
def increment_position(key, dict):
    update_value_in_dicts(key, dict, 'position', search_key_in_dicts(key, dict).get('position') + 1)
    return search_key_in_dicts(key, dict).get('position')
def increment_class_position(key, dict):
    update_value_in_dicts(key, dict, 'class_position', search_key_in_dicts(key, dict).get('class_position') + 1)
    return search_key_in_dicts(key, dict).get('class_position')
    
#loop to run through position, takes a while...
for i in range(0, len(new_data_with_filters_sorted_hour)):
    x = new_data_with_filters_sorted_hour.iloc[i]
    increment_position(x['key'], data_keys_dict)
    new_data_with_filters_sorted_hour.at[i, 'position'] = search_key_in_dicts(x['key'], data_keys_dict).get('position')
```

```python
#gets dict for sorted hour by class
new_data_with_filters_sorted_hour['key_class'] = new_data_with_filters_sorted_hour['key'] + "_" +  new_data_with_filters_sorted_hour['class']
wanted_columns = ['key_class', 'round', 'lap_number']
data_keys_class = new_data_with_filters_sorted_hour.groupby('key_class').mean().reset_index()[wanted_columns].sort_values(['round','lap_number']).reset_index(drop=True)
data_keys_class['key'] = data_keys_class['key_class']
data_keys_class = data_keys_class.drop(['key_class'], axis=1)
data_keys_class['class_position'] = 0
data_keys_class_dict = data_keys_class.to_dict('records')

#loop to run through position, takes a while...
for i in range(0, len(new_data_with_filters_sorted_hour)):
    x = new_data_with_filters_sorted_hour.iloc[i]
    increment_class_position(x['key_class'], data_keys_class_dict)
    new_data_with_filters_sorted_hour.at[i, 'class_position'] = search_key_in_dicts(x['key_class'], data_keys_class_dict).get('class_position')
```

```python
data_sorted_by_position = new_data_with_filters_sorted_hour.sort_values(['season','round', 'lap_number', 'position']).reset_index(drop=True)
#full_data_sorted_by_position

#gets interval overall
data_shift_minus_one = data_sorted_by_position.shift(-1)
data_shift_minus_one['interval'] = data_shift_minus_one['elapsed_ms'] - data_sorted_by_position['elapsed_ms']
negative_interval = data_shift_minus_one[data_shift_minus_one['interval'] < 0]
positive_interval = data_shift_minus_one[data_shift_minus_one['interval'] >= 0]
negative_interval.interval = 0
data_shift_minus_one = negative_interval.append(positive_interval).sort_values(['season','round', 'lap_number', 'position']).reset_index(drop=True)
data_sorted_by_position['interval_ms'] = data_shift_minus_one['interval']
data_sorted_by_position['interval_ms'] = data_sorted_by_position['interval_ms'].shift(1, fill_value=0)
data_sorted_by_position['interval'] = data_sorted_by_position['interval_ms']/1000
data_sorted_by_position
```

```python
#this gets gap for overall
new_data_sorted_by_position = data_sorted_by_position.sort_values(['season','round', 'lap_number', 'position'])
new_data_sorted_by_position['gap'] = new_data_sorted_by_position.groupby('key')['interval'].cumsum()
#this sorts it by class and gets interval and gap for class
new_data_sorted_by_position = new_data_sorted_by_position.sort_values(['season','round', 'lap_number', 'class', 'class_position']).reset_index(drop=True)
new_data_sorted_by_position['class_interval'] = np.where(new_data_sorted_by_position['class_position']!=1, new_data_sorted_by_position['interval'], 0)
new_data_sorted_by_position['class_gap'] = new_data_sorted_by_position.groupby('key_class')['class_interval'].cumsum()
```

```python
#in this one, I'm fixing the time data to be read in as large or seconds, large would be the timedelta object, otherwise our data is now in seconds.
def convert_to_s(x):
    hour = 0
    try:
        hour, minutes, second = x.split(":")
    except ValueError:
        hour = 0
    try: 
        minutes, second = x.split(":")
    except ValueError:
        minutes = 0
    hour_seconds = int(hour)*60*60
    if hour == 0 and minutes == 0:
        return x
    else:
        seconds, ms = second.split(".")
    minute_seconds = (int(minutes)*60 + int(seconds) + int(hour_seconds))*1000 
    total_ms = minute_seconds + int(ms)
    return total_ms/1000

timing_columns_to_fix = ['s1', 's2', 's3']
#all of the x_large data is then converted into into seconds, and then converted into a time_delta variable for ease of displaying data/comparison; this helps for visualizations later
for t in timing_columns_to_fix:
    timing_large_label = t + '_large'
    new_data_sorted_by_position[t] = new_data_sorted_by_position[t].map(str).map(convert_to_s).map(float)
    new_data_sorted_by_position[timing_large_label] = pd.to_timedelta(fia_wec_data[t], unit = 'S')
new_data_sorted_by_position['lap_time'] = pd.to_timedelta(new_data_sorted_by_position['lap_time_s'], unit = 'S')
new_data_sorted_by_position['elapsed_s'] = new_data_sorted_by_position['elapsed_ms']/1000
new_data_sorted_by_position['elapsed'] = pd.to_timedelta(new_data_sorted_by_position['elapsed_s'], unit = 'S')

#dropping key and key_class
new_data_sorted_by_position = new_data_sorted_by_position.drop(['key','key_class'], axis=1)
new_data_sorted_by_position
```

```python
#puts the data to a new csv file
fia_wec_data = fia_wec_data.append(new_data_sorted_by_position).reset_index(drop=True)
fia_wec_data.to_csv("2012-2022_FIA_WEC_FULL_LAP_DATA_new.csv")
```
