```python
fia_wec_data = pd.read_csv("2012-2022_FIA_WEC_FULL_LAP_DATA.csv", index_col = 0)
```

```python
#something we should do is measure the stint number. 

#is 2012 data right?... seems so light
#it is, just wasn't pulilng all of it it seems
fia_wec_data.season = fia_wec_data.season.map(str)
fia_wec_data.groupby('season').mean()
```

```python
#this takes way too long fuck me
#when we do imsa, we need to come up with a plan
#okay so let's just grab a season or two of data so I can test this fucking shit

fia_wec_data_2012 = fia_wec_data[(fia_wec_data['season'] == "2012")].reset_index(drop=True)
#fia_wec_data_2012 = fia_wec_data_2012[(fia_wec_data_2012['class'] == "LMP2")].reset_index(drop=True)
fia_wec_data_2012['key'] = fia_wec_data_2012['season'] + "_" + fia_wec_data_2012['round'].map(int).map(str) + "_" + fia_wec_data_2012['number'].map(int).map(str)
#fia_wec_data_2012['key'] = fia_wec_data_2012['key_team'] + "_" + fia_wec_data_2012['driver_number'].map(int).map(str)
#fia_wec_data_2012_to_2014.season = fia_wec_data_2012_to_2014.season.map(int)
fia_wec_data_2012 = fia_wec_data_2012.sort_values(['season', 'round', 'number', 'lap_number'], axis = 0).reset_index(drop=True)
fia_wec_data_2012.groupby(['round','circuit']).mean()
```

```python
#let's take some time to brain storm this, 'cause I can't have this shit taking this facking long
#we have a few things, actually some scalars we can use instead of names to keep comparing
#we wanna look at the next one, so we do pd.shift -1 I think will get us the next one
#we can use: season, round, number, driver_number, lap_number crossing_finish_line_in_pit(although, we can just check if the driver number changes?)
    #so actually the two conditions are:
        #do we even need the "crossing finish line in pit" line?...
        #any number besides driver number changes - reset to 1
        #driver_number changes, pull from mapping, add one.
wanted_columns = ['season', 'round', 'number', 'driver_number', 'key']
fia_wec_data_pivoted = fia_wec_data_2012.groupby(wanted_columns).mean().reset_index()[wanted_columns]
fia_wec_data_pivoted['team_stint_no'] = 0
```

```python
fia_wec_data_2012['driver_stint_no'] = 0

#we take the next lap: next = df shifted by 1 row (starts from lap 2)
fia_wec_data_2012_next = fia_wec_data_2012.shift(-1).fillna(0)

#check where the driver changes, different from "crossing line in pit" since we know that the driver changes.
fia_wec_data_2012['change_team'] = np.where(fia_wec_data_2012['key'] == fia_wec_data_2012_next['key'], 0, 1)
fia_wec_data_2012['change_team'] = fia_wec_data_2012['change_team'].shift(1).fillna(0)

#check where the stint changes, "crossing line in pit", but we have to increment current driver stint as well
fia_wec_data_2012['change_stint'] = np.where(fia_wec_data_2012['crossing_finish_line_in_pit'] == "B", 1, 0)
fia_wec_data_2012['change_stint'] = fia_wec_data_2012['change_stint'].shift(1).fillna(0)

wanted_columns = ['key', 'team_stint_no']
fia_wec_data_pivoted_wc = fia_wec_data_pivoted[wanted_columns]
fia_wec_data_pivoted_dict = fia_wec_data_pivoted_wc.to_dict('records')
```

```python
def search_key_in_dicts(key, dict):
    for d in dict:
        if d['key'] == key:
            return d
    return None
def update_value_in_dicts(key, dict, col, value):
    dict_key = search_key_in_dicts(key, dict)
    dict_key.update({col : value})
def increment_driver_stint_value(key, dict):
    update_value_in_dicts(key, dict, 'driver_stint_no', search_key_in_dicts(key, dict).get('driver_stint_no') + 1)
    return search_key_in_dicts(key, dict).get('driver_stint_no')
def increment_team_stint_value(key, dict):
    update_value_in_dicts(key, dict, 'team_stint_no', search_key_in_dicts(key, dict).get('team_stint_no') + 1)
    return search_key_in_dicts(key, dict).get('team_stint_no')
def determine_change_stint_no(change_driver, change_stint):
    if change_driver == 1 or change_stint == 1:
        return 1
    else:
        return 0

key = fia_wec_data_2012.at[0, 'key']
update_value_in_dicts(key, fia_wec_data_pivoted_dict, 'team_stint_no', 1)
```

```python
#working case for 2012 data, run time is 7.5s
fia_wec_data_2012['team_stint_no'] = 0
for i in range(0,len(fia_wec_data_2012)):
    x = fia_wec_data_2012.iloc[i]
    if x['change_stint'] == 1 or x['change_team'] == 1:
        increment_team_stint_value(x.key, fia_wec_data_pivoted_dict)
    fia_wec_data_2012.at[x.name, 'team_stint_no'] = (search_key_in_dicts(x.key, fia_wec_data_pivoted_dict).get('team_stint_no'))
```
