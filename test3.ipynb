```python
import pandas as pd
from tabula.io import read_pdf
```

```python
#here we're using tabula to read the pdf and grab the table from it.
test_1 = read_pdf('sebring_entry_list.pdf', pandas_options={'header':1})[0]
test_1
```

```python
#manually fixed this, basically the data is unclean. I can explain what I did, but I think through coding it would have been less efficient, since it's such a small set.
#changed the titles to reflect better what we're looking for,
#added a comma and fixed line 9
#removed the other useless rows at the bottom

drivers = pd.read_csv('entries.csv')
drivers
```

```python
#the drivers are stacked in columns driver_1 to driver_4, so we have to unstack them
driver_columns = ['driver', "license"]
drivers_with_licenses = pd.DataFrame(columns = driver_columns)
#essentially this loop takes what I put for driver_x and license_x and combines them into drivers_with_licenses
for i in range(1,5):
    driver_col = "driver_" + str(i)
    license_col = "license_" + str(i)
    wanted_columns = [driver_col, license_col]
    
    iter_drivers = drivers[wanted_columns]
    iter_drivers.columns = driver_columns
    drivers_with_licenses = drivers_with_licenses.append(iter_drivers)

drivers_with_licenses
```

```python
#still some more work to do. There's NA's in licenses, let's get rid of those first.
drivers_with_licenses =  drivers_with_licenses[drivers_with_licenses['license'].notna()]
drivers_with_licenses
```

```python
#let's also compare to what we currently have for driver_name in the sebring pdf
sebring_laps = pd.read_csv('1000_Sebring_Laps.csv', delimiter = ";")
sebring_laps.groupby("DRIVER_NAME").mean()
```

```python
#we have to fix this name, could do it by creating a function that takes the driver names and takes the nationality out of the name
def fix_driver_names(x):
    name_array = x.split(" ")
    name_array = name_array[:-1]
    name = " ".join(name_array)
    return name

#so let's apply it to our df
drivers_with_licenses.driver = drivers_with_licenses.driver.apply(fix_driver_names)
drivers_with_licenses
```

```python
#we got the names free! but...
sebring_laps.DRIVER_NAME
```

```python
#as we can kinda see, it doesn't really match the previous one.
#we can fix this by making them all lower or uppercase, let's go with lower
sebring_laps_lower = sebring_laps.copy()
sebring_laps_lower['DRIVER_NAME'] = sebring_laps['DRIVER_NAME'].str.lower()
sebring_laps_lower
```

```python
#doing the same on the licenses
drivers_with_licenses['driver'] = drivers_with_licenses['driver'].str.lower()
drivers_with_licenses
```

```python
#now let's see if this combines them!
laps_with_licenses = pd.merge(left = sebring_laps_lower, right = drivers_with_licenses, left_on = "DRIVER_NAME", right_on = "driver", how = "outer")
laps_with_licenses
```

```python
#since we did an outer join, we can see that there are people that got omitted from both dfs
laps_with_licenses_na = laps_with_licenses[laps_with_licenses['CLASS'].isna()]
laps_with_licenses_na
```

```python
#some of them didn't race, like Mike Conway or Alex Brundle.
#But some of them do exist in the df, specifically Marco SØrensen, Ferdinand Habsburg, and Oliver Millroy (misspelled as Olliver)
sebring_laps_omitted = sebring_laps[(sebring_laps['DRIVER_NAME'] == 'Marco SØRENSEN') | (sebring_laps['DRIVER_NAME'] == "Ferdinand HABSBURG") | (sebring_laps['DRIVER_NAME'] == "Oliver MILLROY")]
sebring_laps_omitted
```

```python
#let's grab them from the DF
omitted_wanted_columns = ["DRIVER_NAME"]
laps_with_licenses_na = laps_with_licenses[laps_with_licenses['license'].isna()]
laps_with_licenses_na_drivers = laps_with_licenses_na[omitted_wanted_columns]
laps_with_licenses_na_drivers["driver"] = laps_with_licenses_na["DRIVER_NAME"].str.lower()
laps_with_licenses_na_drivers = laps_with_licenses_na_drivers.drop_duplicates().reset_index(drop = True)
laps_with_licenses_na_drivers = laps_with_licenses_na_drivers.drop("DRIVER_NAME", axis = 1)
laps_with_licenses_na_drivers
```

```python
#fabio scherer and rene binder aren't on the fia list for some reason
#just a google search shows rene is a silver and fabio is a gold

omitted_drivers_license = ["P", "G", "G", "S", "S"]
#now that we have this, we can append this to the driver_licenses df
laps_with_licenses_na_drivers['license'] = omitted_drivers_license
laps_with_licenses_na_drivers
```

```python
#so let's add them to the drivers_with_licenses df
drivers_with_licenses = drivers_with_licenses.append(laps_with_licenses_na_drivers).reset_index(drop=True)
drivers_with_licenses
```

```python
#let's try combining them again!
laps_with_licenses = pd.merge(left = sebring_laps_lower, right = drivers_with_licenses, left_on = "DRIVER_NAME", right_on = "driver", how = "outer")
laps_with_licenses
```

```python
#any nas?
laps_with_licenses_na = laps_with_licenses[laps_with_licenses['license'].isna()]
laps_with_licenses_na 
```

```python
#okay cool! let's drop the column and bottom nas
laps_with_licenses = laps_with_licenses[laps_with_licenses["TEAM"].notna()]

#and also drop the driver column 
laps_with_licenses = laps_with_licenses.drop("driver", axis=1)
laps_with_licenses
```

```python
#nooooow we can do my analysis, so let's save this as a csv
laps_with_licenses.to_csv('sebring_data_with_licenses.csv')
```
