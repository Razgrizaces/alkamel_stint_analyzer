```python
import pandas as pd
import numpy as np
```

```python
fia_wec_data = pd.read_csv('2012-2022_FIA_WEC_FULL_LAP_DATA.csv', index_col = 0)
```

```python
fia_wec_data['season'] = fia_wec_data['season'].map(str)
#break it into gte, p2 and p1
fia_wec_data_gte = fia_wec_data[fia_wec_data['class'].str.contains("LMGTE")].reset_index(drop=True)
fia_wec_data_lmp2 = fia_wec_data[fia_wec_data['class'].str.contains("LMP2")].reset_index(drop=True)
fia_wec_data_lmp1 = fia_wec_data[fia_wec_data['class'].str.contains("LMP1")|fia_wec_data['class'].str.contains("HYPERCAR")].reset_index(drop=True)
fia_wec_data_misc = fia_wec_data[fia_wec_data['class'].str.contains("INNOVATIVE CAR")|fia_wec_data['class'].str.contains("CDNT")]
fia_wec_data_misc.reset_index(drop=True)
```

```python
fia_wec_data.groupby("class").mean()
```

```python
#oh this might be easier if I just do this:
gte_vehicles = fia_wec_data_gte.groupby(["season", "vehicle"]).mean().reset_index()
gte_vehicles['vehicle_actual'] = np.nan
wanted_columns = ['vehicle', 'season','vehicle_actual']
gte_vehicles = gte_vehicles[wanted_columns]
#gte_vehicles.to_csv("gtes.csv")
```

```python
gte_vehicles = pd.read_csv("gtes.csv", index_col=0)
gte_vehicles['vehicle_season'] = gte_vehicles['vehicle'] + "_" + gte_vehicles['season']
wanted_columns = ['vehicle_actual', 'vehicle_season']
gte_vehicles_fixed = gte_vehicles[wanted_columns]
gte_vehicles_fixed
```

```python
fia_wec_data_gte['vehicle_season'] = fia_wec_data_gte['vehicle'] + "_" + fia_wec_data_gte['season']
fia_wec_data_gte
```

```python
fia_wec_data_gte_fixed = pd.merge(left = fia_wec_data_gte, right = gte_vehicles_fixed, on = "vehicle_season", how = "left")
#fia_wec_data_gte_fixed['vehicle'] = fia_wec_data_gte_fixed['vehicle_actual']
fia_wec_data_gte_fixed = fia_wec_data_gte_fixed.drop(['vehicle_season', 'vehicle_actual'], axis=1)
fia_wec_data_gte_fixed
```

```python
gte_vehicles_fixed = pd.read_csv("gtes.csv", index_col=0)
wanted_columns = ['vehicle_actual', 'season']
gte_vehicles_fixed = gte_vehicles_fixed[wanted_columns]
gte_vehicles_fixed = gte_vehicles_fixed[(gte_vehicles_fixed['season'] == "2021")|(gte_vehicles_fixed['season'] == "2018-2019")|(gte_vehicles_fixed['season'] == "2016")].drop_duplicates().reset_index(drop=True)
gte_vehicles_fixed['manufacturer'] = gte_vehicles_fixed['vehicle_actual'].str.split(expand = True)[0]
gte_vehicles_fixed['man_season'] = gte_vehicles_fixed['manufacturer'] + "_" + gte_vehicles_fixed['season'] 
gte_vehicles_fixed = gte_vehicles_fixed.groupby(['man_season', 'vehicle_actual']).mean().drop_duplicates().reset_index()
gte_vehicles_fixed
```

```python
gte_data_missing = fia_wec_data_gte_fixed[fia_wec_data_gte_fixed['vehicle'].isna()]
gte_data_notmissing = fia_wec_data_gte_fixed[fia_wec_data_gte_fixed['vehicle'].notna()]

gte_missing_vehicles = pd.read_csv("gtes_missing.csv", index_col=0)
wanted_columns = ["vehicle", "team_no"]

gte_missing_vehicles = gte_missing_vehicles[wanted_columns]
gte_data_missing = gte_data_missing.drop("vehicle", axis=1)
gte_data_missing_with_vehicles = pd.merge(left=gte_data_missing, right=gte_missing_vehicles, on = "team_no", how = "outer")
gte_data_missing_with_vehicles
```

```python
fia_wec_data_gte_fixed = gte_data_notmissing.append(gte_data_missing_with_vehicles).reset_index(drop=True)
fia_wec_data_gte_fixed
```

```python
fia_wec_data_gte_without_keating = fia_wec_data_gte_fixed[fia_wec_data_gte_fixed['team'] != 'Keating Motorsports']
fia_wec_data_gte_with_keating = fia_wec_data_gte_fixed[fia_wec_data_gte_fixed['team'] == 'Keating Motorsports']
fia_wec_data_gte_with_keating_2018 = fia_wec_data_gte_with_keating[fia_wec_data_gte_with_keating['circuit'] == 'LE_MANS_2018']
fia_wec_data_gte_with_keating_2019 = fia_wec_data_gte_with_keating[fia_wec_data_gte_with_keating['circuit'] == 'LE_MANS_2019']
fia_wec_data_gte_with_keating_2018['vehicle'] = "Ferrari 488 GTE"

fia_wec_data_with_keating_fixed = fia_wec_data_gte_with_keating_2018.append(fia_wec_data_gte_with_keating_2019)

fia_wec_data_gte_fixed = fia_wec_data_gte_without_keating.append(fia_wec_data_with_keating_fixed).reset_index(drop=True)
fia_wec_data_gte_fixed[fia_wec_data_gte_fixed['vehicle'].isna()]
fia_wec_data_gte_fixed['engine'] = fia_wec_data_gte_fixed['manufacturer']
fia_wec_data_gte_fixed
```

```python
lmp2_2018_nas_with_classification = pd.merge(left = lmp2_2018_nas, right=lmp2_2018_classification, on = 'number', how = 'outer').dropna(axis=1, how="all")
```

```python
import pandas as pd

#why do we need these again?... i mean i guess it's nice to have

wanted_columns = ['vehicle']

lmp1_vehicles = pd.read_csv("lmp1s.csv", index_col=0)
lmp2_vehicles = pd.read_csv("lmp2s.csv", index_col=0)

lmp1_vehicles['vehicle'] = lmp1_vehicles['vehicle_actual']

lmp2_vehicles = lmp2_vehicles[wanted_columns]
lmp1_vehicles = lmp1_vehicles[wanted_columns]
lmp1_vehicles = lmp1_vehicles.drop_duplicates().reset_index(drop=True)
lmp2_vehicles = lmp2_vehicles.drop_duplicates().reset_index(drop=True)

lmp_vehicles = lmp1_vehicles.append(lmp2_vehicles).reset_index(drop=True)

lmp_vehicles_with_engine = lmp_vehicles.vehicle.str.split(" - ",expand = True).drop_duplicates().reset_index(drop=True)
lmp_vehicles_with_engine['vehicle_actual']  = lmp_vehicles_with_engine[0] + " - " +  lmp_vehicles_with_engine[1] 
lmp_vehicles_with_engine['vehicle'] = lmp_vehicles_with_engine[0]
lmp_vehicles_with_engine['engine'] = lmp_vehicles_with_engine[1]

lmp_vehicles_with_engine = lmp_vehicles_with_engine.drop([0,1],axis=1)
lmp_vehicles_with_engine
```

```python
lmp1_vehicles = pd.read_csv("lmp1s.csv", index_col=0)
lmp2_vehicles = pd.read_csv("lmp2s.csv", index_col=0)

lmp1_vehicles = lmp1_vehicles.drop(["manufacturer", "season"], axis=1).drop_duplicates().reset_index(drop=True)
lmp2_vehicles = lmp2_vehicles.drop(["number", "season"], axis=1).drop_duplicates().reset_index(drop=True)

lmp2_vehicles['vehicle_actual'] = lmp2_vehicles['vehicle']
lmp_vehicles = lmp1_vehicles.append(lmp2_vehicles).drop_duplicates().reset_index(drop=True)

lmp_vehicles
```
