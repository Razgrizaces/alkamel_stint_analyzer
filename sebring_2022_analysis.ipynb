```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#these will allow us to plot the data shown
from matplotlib import pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
#load the data we're going to be using
fia_wec_data = pd.read_csv('2012-2022_FIA_WEC_FULL_LAP_DATA.csv', index_col = 0)
fia_wec_data.groupby('season').mean()
```

```python
#for some reason the season doesn't map properly sometimes (some as ints, some as string), so we have to map it. 
fia_wec_data['season'] = fia_wec_data['season'].map(str)
fia_wec_data.groupby('season').mean()
```

```python
#now we can grab sebring_2022 data
sebring_2022 = fia_wec_data[fia_wec_data['season'] == "2022"].reset_index(drop=True)
sebring_2022
```

```python
#we can group by the team_no to get all the teams + the number of the team
sebring_2022.groupby("team_no").mean()
```

```python
#let's pull out the lmp2 lap times
sebring_2022_lmp2_data = sebring_2022[sebring_2022['class'] == "LMP2"].reset_index(drop=True)
sebring_2022_lmp2_data
```

```python
#now let's try to plot them
sns.set(rc={'figure.figsize':(35,12)})
plt.style.use('dark_background')
sns.boxplot(data = sebring_2022_lmp2_data, x = 'team_no', y = 'lap_time_s', palette = 'bright')
```

```python
#we realize that there's a lot of outliers, let's try to take the 115% time
#this 115% of the minimum time, used in F1 qualifying (107% instead of 115%), it shows a more representative pace lap
#some of the outliers are due to pit stops, passing cars on track, damage, etc.
cutoff_time = min(sebring_2022_lmp2_data.lap_time_s)* 1.15
sebring_2022_lmp2_data_with_cutoff = sebring_2022_lmp2_data[sebring_2022_lmp2_data['lap_time_s']<cutoff_time].reset_index(drop=True)
sebring_2022_lmp2_data_with_cutoff
```

```python
#this is a sample of some of the plots you can make with this data set
sns.set(font_scale = 1.5, rc={'figure.figsize':(24,12)})
plt.style.use('dark_background')
ax = sns.boxplot(data = sebring_2022_lmp2_data_with_cutoff, x = 'team_no', y = 'lap_time_s', palette = 'bright')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 25,wrap = True)
ax.set_title("LMP2 Lap Time Box Plots by Team")
ax.set_xlabel("LMP2 Team")
ax.set_ylabel("Lap Time (s)")
```

```python
#let's combine the gte field
sebring_2022_gte = sebring_2022[(sebring_2022['class'] == "LMGTE Pro") | (sebring_2022['class'] == "LMGTE Am")]
sebring_2022_gte = sebring_2022_gte.reset_index(drop = True)

#and if we want to analyze gte_pro only we can just filter for it
sebring_2022_gte_pro = sebring_2022_gte[sebring_2022_gte['class'] == "LMGTE Pro"]
sebring_2022_gte_pro.groupby("team_no").mean()
```

```python
#also applying the 107 rule here
cutoff_time_gte = sebring_2022_gte_pro.lap_time_ms.min() * 1.07
sebring_2022_gte_pro_107_time = sebring_2022_gte_pro[sebring_2022_gte_pro['lap_time_ms'] < cutoff_time_gte]
sebring_2022_gte_pro_107_time = sebring_2022_gte_pro_107_time.reset_index(drop = True)
sebring_2022_gte_pro_107_time
```

```python
#we can also do something like this, plotting the gte pro lap times by team with colors based on manufacturer
man_colors = {"Ferrari":"r", 'Aston Martin':"g", "Porsche":'white', 'Chevrolet':"yellow"}
sns.set(font_scale = 1.5, rc={'figure.figsize':(12,8)})
plt.style.use('dark_background')
ax2 = sns.boxplot(data = sebring_2022_gte_pro_107_time, x = 'team_no', y = 'lap_time_s', palette = man_colors, hue = "manufacturer")
ax2.set_xticklabels(ax2.get_xticklabels(), rotation = 25, wrap = True)
ax2.set_xlabel("Team")
ax2.set_ylabel("Lap Time (s)")
ax2.set_title("GTE Pro Lap Time by Team")
```

```python
#now we can see that the #92 porsche (which ended up winning the race) had consistent and faster lap times (lower is faster)
#we can even go a step further and group it by driver name instead of team. 
sns.set(font_scale = 1.25, rc={'figure.figsize':(12,8)})
plt.style.use('dark_background')
ax3 = sns.boxplot(data = sebring_2022_gte_pro_107_time, x = 'driver_name', y = 'lap_time_s', palette = man_colors, hue = "manufacturer")
ax3.set_xticklabels(ax3.get_xticklabels(), rotation = 30,wrap = True)
ax3.set_xlabel("Driver")
ax3.set_ylabel("Lap Time (s)")
ax3.set_title("GTE PRO Lap Times by Driver")
```

```python
#if we want to look overall at the gtes
#increasing the cutoff time a bit as well since some ams get cut off
cutoff_time_gte = min(sebring_2022_gte_pro.lap_time_s) * 1.10
sebring_2022_gte_107_time = sebring_2022_gte[sebring_2022_gte['lap_time_s'] < cutoff_time_gte]
sebring_2022_gte_107_time = sebring_2022_gte_107_time.reset_index(drop = True)
```

```python
sns.set(font_scale = 1.5, style = 'darkgrid', rc={'figure.figsize':(24,10)})
plt.style.use('dark_background')

gte_colors = {"LMGTE Am": 'orange', "LMGTE Pro" : "green"}
ax4 = sns.boxplot(data = sebring_2022_gte_107_time, x = 'team_no', y = 'lap_time_s', palette = gte_colors, hue = "class")
ax4.set_xticklabels(ax4.get_xticklabels(), rotation = 25,wrap = True)
ax4.set_xlabel("Team")
ax4.set_ylabel("Lap Time (s)")
ax4.set_title("GTE Lap Time by Team by Class")
```

```python
#here we're pulilng the hypercar data and also applying the 107% rule to it as well
sebring_2022_hypercar = sebring_2022[sebring_2022['class'] == "HYPERCAR"].reset_index(drop=True)
hypercar_cutoff = sebring_2022_hypercar.lap_time_ms.min() * 1.07
sebring_2022_hypercar_cutoff = sebring_2022_hypercar[sebring_2022_hypercar['lap_time_ms'] < hypercar_cutoff].reset_index(drop=True)
sebring_2022_hypercar_cutoff
```

```python
#here's the hypercar lap times by team/driver
hypercar_colors = {"Alpine ELF Team #36": 'b', "Glickenhaus Racing #708": "r", "Toyota Gazoo Racing #7": "gray", "Toyota Gazoo Racing #8": 'white'}

sns.set(font_scale = 1.5, style = 'darkgrid', rc={'figure.figsize':(22,10)})
plt.style.use('dark_background')
fig2, axs2 = plt.subplots(ncols = 2)

sns.boxplot(data = sebring_2022_hypercar_cutoff, x = 'team_no', y = 'lap_time_s', palette = hypercar_colors, ax=axs2[0])
axs2[0].set_xticklabels(axs2[0].get_xticklabels(), rotation = 30,wrap = True)
axs2[0].set_title("Hypercar Lap Times by Team")
axs2[0].set_xlabel("Team")
axs2[0].set_ylabel("Lap Time (s)")
sns.boxplot(data = sebring_2022_hypercar_cutoff, x = 'driver_name', y = 'lap_time_s', palette = hypercar_colors, hue = "team_no", ax=axs2[1])
axs2[1].set_xticklabels(axs2[1].get_xticklabels(), rotation = 30,wrap = True)
axs2[1].set_title("Hypercar Lap Times by Driver")
axs2[1].set_xlabel("Team")
axs2[1].set_ylabel("Lap Time (s)")
```

```python
#and the gte pro lap times by team/driver
gte_pro_team_colors = {"AF Corse #51": "red", "AF Corse #52": "orange", "Corvette Racing #64": "yellow", "Porsche GT Team #91":"white", "Porsche GT Team #92":"gray" }

fig3, axs3 = plt.subplots(ncols = 2)

sns.set(font_scale = 1.25, rc={'figure.figsize':(24,8)})
plt.style.use('dark_background')
sns.boxplot(data = sebring_2022_gte_pro_107_time, x = 'team_no', y = 'lap_time_s', palette = gte_pro_team_colors, ax=axs3[0])
axs3[0].set_xticklabels(axs3[0].get_xticklabels(), rotation = 25, wrap = True)
axs3[0].set_xlabel("Team")
axs3[0].set_ylabel("Lap Time (s)")
axs3[0].set_title("GTE Pro Lap Time by Team")

sns.boxplot(data = sebring_2022_gte_pro_107_time, x = 'driver_name', y = 'lap_time_s', palette = gte_pro_team_colors, hue = "team_no", ax=axs3[1])
axs3[1].set_xticklabels(ax3.get_xticklabels(), rotation = 30,wrap = True)
axs3[1].set_xlabel("Driver")
axs3[1].set_ylabel("Lap Time (s)")
axs3[1].set_title("GTE PRO Lap Times by Driver")
```
