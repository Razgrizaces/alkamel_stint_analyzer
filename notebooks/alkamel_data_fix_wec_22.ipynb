```python
import pandas as pd
import numpy as np
alkamel_data = pd.read_csv('alkamel_data_full.csv')
```

```python
alkamel_data_nas = alkamel_data[alkamel_data['team'].isna()]
alkamel_data_nas
```

```python
alkamel_data_nas.groupby(['championship', 'season' , 'round', 'session']).count()
```

```python
alkamel_data_fiawec = alkamel_data[alkamel_data['championship'] == 'FIAWEC']
alkamel_data_fiawec['team_key'] = alkamel_data_fiawec['season'].map(str) + "_" + alkamel_data_fiawec['championship'] + '_' \
    + alkamel_data_fiawec['number'].map(str)
alkamel_data_fiawec.groupby(['season', 'championship', 'round', 'number', 'team', 'manufacturer', 'vehicle']).count().to_csv('fia_wec_teams.csv')
```

```python
alkamel_data_fiawec_2013 = alkamel_data_fiawec[alkamel_data_fiawec['season'] == 2013]
alkamel_data_fiawec_2013_race = alkamel_data_fiawec_2013[alkamel_data_fiawec_2013['session_type'] == 'RACE']
alkamel_data_fiawec_2013_race_no_r4 = alkamel_data_fiawec_2013_race[alkamel_data_fiawec_2013_race['round'] != 4]
wanted_columns = ['team_key','team', 'vehicle']
alkamel_fiawec_2013_teams = alkamel_data_fiawec_2013_race_no_r4.groupby(['team_key', 'round', 'team', 'vehicle']).count().reset_index()
alkamel_fiawec_2013_teams = alkamel_fiawec_2013_teams[wanted_columns].drop_duplicates()
alkamel_fiawec_2013_teams = alkamel_fiawec_2013_teams.drop([132])
alkamel_fiawec_2013_teams
```

```python
alkamel_data_fiawec_2013_qp = alkamel_data_fiawec_2013[alkamel_data_fiawec_2013['session_type'] == 'QUALIFYING']
alkamel_data_fiawec_2013_qp_na = alkamel_data_fiawec_2013_qp[alkamel_data_fiawec_2013_qp['team'].isna()]
```

```python
alkamel_data_2013_qp_fix = pd.merge(alkamel_data_fiawec_2013_qp_na, alkamel_fiawec_2013_teams, on = 'team_key')
alkamel_data_fiawec_2013_qp_na.groupby('team_key').count() - alkamel_data_2013_qp_fix.groupby('team_key').count()

#fix team 
alkamel_data_2013_qp_fix['team'] = alkamel_data_2013_qp_fix['team_y']
alkamel_data_2013_qp_fix['vehicle'] = alkamel_data_2013_qp_fix['vehicle_y']
alkamel_data_2013_qp_fix = alkamel_data_2013_qp_fix.drop(['vehicle_x', 'vehicle_y', 'team_x', 'team_y'],axis=1)
alkamel_data_2013_qp_fix
```

```python
#build the qp with r4
alkamel_data_fiawec_2013_qp_r4 = alkamel_data_fiawec_2013[alkamel_data_fiawec_2013['round'] == 4]
alkamel_data_fiawec_2013_qp_r4 = alkamel_data_fiawec_2013_qp_r4[alkamel_data_fiawec_2013_qp_r4['session_type'] == 'QUALIFYING']

alkamel_data_fiawec_2013_full_qp = pd.concat([alkamel_data_2013_qp_fix, alkamel_data_fiawec_2013_qp_r4])
#build 2013 with qp
alkamel_data_fiawec_2013_no_qp = alkamel_data_fiawec_2013[alkamel_data_fiawec_2013['session_type'] != 'QUALIFYING']
alkamel_data_fiawec_2013_fix = pd.concat([alkamel_data_fiawec_2013_no_qp, alkamel_data_fiawec_2013_full_qp])
```

```python
alkamel_data_fiawec['team_key'] = alkamel_data_fiawec['season'].map(str) + "_" + alkamel_data_fiawec['championship'] + '_' \
    + alkamel_data_fiawec['round'].map(str) +"_" + alkamel_data_fiawec['number'].map(str)
alkamel_data_fiawec_2014 = alkamel_data_fiawec[alkamel_data_fiawec['season'] == 2014]
alkamel_data_fiawec_2014_race = alkamel_data_fiawec_2014[alkamel_data_fiawec_2014['session_type'] == 'RACE']
alkamel_data_fiawec_2014_race_r6 = alkamel_data_fiawec_2014_race[alkamel_data_fiawec_2014_race['round'] <= 6]
alkamel_data_fiawec_2014_race_r6
wanted_columns = ['team_key','team', 'vehicle']
alkamel_fiawec_2014_teams = alkamel_data_fiawec_2014_race_r6.groupby(['team_key', 'round', 'team', 'vehicle']).count().reset_index()
alkamel_fiawec_2014_teams = alkamel_fiawec_2014_teams[wanted_columns].drop_duplicates()
alkamel_fiawec_2014_teams
```

```python
alkamel_data_fiawec_2014_qp = alkamel_data_fiawec_2014[alkamel_data_fiawec_2014['session_type'] == 'QUALIFYING']
alkamel_data_fiawec_2014_qp
```

```python
alkamel_data_fiawec_2014_qp_na = alkamel_data_fiawec_2014_qp[alkamel_data_fiawec_2014_qp['team'].isna()]
alkamel_data_fiawec_2014_qp_na
```

```python
alkamel_data_2014_qp_fix = pd.merge(alkamel_data_fiawec_2014_qp_na, alkamel_fiawec_2014_teams, on = 'team_key')
alkamel_data_2014_qp_fix
```

```python
#fix team for 2014 now
alkamel_data_2014_qp_fix['team'] = alkamel_data_2014_qp_fix['team_y']
alkamel_data_2014_qp_fix['vehicle'] = alkamel_data_2014_qp_fix['vehicle_y']
alkamel_data_2014_qp_fix = alkamel_data_2014_qp_fix.drop(['vehicle_x', 'vehicle_y', 'team_x', 'team_y'],axis=1)
alkamel_data_2014_qp_fix
```

```python
alkamel_data_fiawec_2014_qp_rest = alkamel_data_fiawec_2014[(alkamel_data_fiawec_2014['round']!=2)&(alkamel_data_fiawec_2014['round']!=3)&(alkamel_data_fiawec_2014['round']!=5)&(alkamel_data_fiawec_2014['round']!=6)]
alkamel_data_fiawec_2014_qp_rest = alkamel_data_fiawec_2014_qp_rest[alkamel_data_fiawec_2014_qp_rest['session_type'] == 'QUALIFYING']

alkamel_data_fiawec_2014_full_qp = pd.concat([alkamel_data_2014_qp_fix, alkamel_data_fiawec_2014_qp_rest])
len(alkamel_data_fiawec_2014_full_qp) - len(alkamel_data_fiawec_2014_qp)
```

```python
#build 2014 with qp
alkamel_data_fiawec_2014_no_qp = alkamel_data_fiawec_2014[alkamel_data_fiawec_2014['session_type'] != 'QUALIFYING']
alkamel_data_fiawec_2014_fix = pd.concat([alkamel_data_fiawec_2014_no_qp, alkamel_data_fiawec_2014_full_qp])
len(alkamel_data_fiawec_2014) - len(alkamel_data_fiawec_2014_fix)
```

```python
alkamel_fiawec_2012_r1_teams = pd.read_csv('2012_r1_teams.csv', index_col=0)
alkamel_data_2012_r1_rest_fix = pd.merge(alkamel_data_fiawec_2012_r1_not_race, alkamel_fiawec_2012_r1_teams, on = 'team_key')
alkamel_data_2012_r1_rest_fix['team'] = alkamel_data_2012_r1_rest_fix['team_y']
alkamel_data_2012_r1_rest_fix['vehicle'] = alkamel_data_2012_r1_rest_fix['vehicle_y']
alkamel_data_2012_r1_rest_fix = alkamel_data_2012_r1_rest_fix.drop(['vehicle_x', 'vehicle_y', 'team_x', 'team_y'],axis=1)
alkamel_data_2012_r1_rest_fix
alkamel_data_2012_r1_rest_fix
```

```python
alkamel_data_2012_r1 = alkamel_data_fiawec_2012[alkamel_data_fiawec_2012['round']==1]
alkamel_data_2012_no_r1 = alkamel_data_fiawec_2012[alkamel_data_fiawec_2012['round']!=1]
alkamel_data_2012_r1_fix = pd.concat([alkamel_data_fiawec_2012_r1_race,alkamel_data_2012_r1_rest_fix])
alkamel_data_fiawec_2012_fix = pd.concat([alkamel_data_2012_no_r1, alkamel_data_2012_r1_fix])
len(alkamel_data_fiawec_2012) - len(alkamel_data_fiawec_2012_fix)
```
