```python
import numpy as np
import pandas as pd
```

```python
#this is to help me create a new pipeline for the fia_wec dataset. 
#maybe it'll help me also make the new dataset work better? not sure XD

fia_wec_data = pd.read_csv("2012-2022_FIA_WEC_FULL_LAP_DATA_old.csv", index_col=0)
fia_wec_data
```

```python
fia_wec_data_2022 = fia_wec_data[fia_wec_data['season'] == "2022"].reset_index(drop=True)
fia_wec_data_2022
```

```python
#reading the new file in
new_data = pd.read_csv('data/2022_SPA_FRANCORCHAMPS.csv', index_col=0)
```

```python
#fixing the columns from the new file
new_data = new_data.dropna(axis=1, how="all")
new_data.columns = new_data.columns.str.lower().str.strip()
new_data.columns
```

```python
#checking which columns we need to add
fia_wec_data_2022.columns
```

```python
#this adds the mapping for the vehicles for that season
wanted_columns = ['number', 'vehicle','engine', 'season']
fia_wec_vehicles = fia_wec_data.groupby(['number', 'vehicle', 'engine', 'season']).mean().reset_index()
fia_wec_engine_mapping = fia_wec_vehicles[wanted_columns].drop_duplicates()
fia_wec_engine_mapping.to_csv("fia_wec_engine_mapping.csv")
```

```python
#only getting the engine/vec map for 2022
fia_wec_veh_to_eng_2022 = fia_wec_engine_mapping[fia_wec_engine_mapping['season'] == "2022"].reset_index(drop=True)
fia_wec_veh_to_eng_2022 = fia_wec_veh_to_eng_2022.drop('season', axis=1)
```

```python
#we need: 'team_no', 'lap_time_ms', 'lap_time_s', 'engine', 'driver_stint_no', 'driver_stint', 'team_stint_no', 'team_stint'
def convert_to_ms(x):
    minutes, second = x.split(":")
    seconds, ms = second.split(".")
    minute_seconds = (int(minutes)*60 + int(seconds))*1000
    total_ms = minute_seconds + int(ms)
    return total_ms

#team_no
new_data['team_no'] = new_data['team'] + " #" + new_data['number'].map(int).map(str)

#lap_time_ms + lap_time_s
#we can then apply this function to every single lap time in the df, and convert everything to ms.
new_data['lap_time_ms'] = new_data['lap_time'].apply(convert_to_ms)
#and then let's make a seconds column to plot it neatly later.
new_data['lap_time_s'] = new_data['lap_time_ms']/1000
```

```python
#vehicle and engine
new_data = pd.merge(new_data, fia_wec_veh_to_eng_2022, on = 'number')
```

```python
#getting team and driver stint nos

#we have two different keys for driver and team, this one is for team
new_data['key'] = new_data['season'].map(int).map(str) + "_" + new_data['round'].map(int).map(str) + "_" + new_data['number'].map(int).map(str)

#we take the next lap: next = df shifted by 1 row (starts from lap 2)
new_data_next = new_data.shift(-1).fillna(0)

#check where the driver changes, different from "crossing line in pit" since we know that the driver changes.
new_data['change_driver'] = np.where(new_data['key'] == new_data_next['key'], 0, 1)
new_data['change_driver'] = new_data['change_driver'].shift(1).fillna(0)

#check where the team changes
new_data['change_team'] = np.where(new_data['number'] == new_data_next['number'], 0, 1)
new_data['change_team'] = new_data['change_team'].shift(1).fillna(0)

new_data['change_stint'] = np.where(new_data['crossing_finish_line_in_pit'] == "B", 1, 0)
new_data['change_stint'] = new_data['change_stint'].shift(1).fillna(0)

#here we're creating a dictionary for the keys by pivoting it by these columns
wanted_columns = ['key']
new_data_pivoted = new_data.groupby(wanted_columns).mean().reset_index()[wanted_columns]
new_data_pivoted['team_stint_no'] = 0

wanted_columns = ['key', 'team_stint_no']
new_data_pivoted_wc = new_data_pivoted[wanted_columns]
new_data_pivoted_dict = new_data_pivoted_wc.to_dict('records')

def search_key_in_dicts(key, dict):
    for d in dict:
        if d['key'] == key:
            return d
    return None
def update_value_in_dicts(key, dict, col, value):
    dict_key = search_key_in_dicts(key, dict)
    dict_key.update({col : value})
def increment_driver_stint_value(key, dict):
    update_value_in_dicts(key, dict, 'driver_stint_no', search_key_in_dicts(key, dict).get('driver_stint_no') + 1)
    return search_key_in_dicts(key, new_data_pivoted_dict).get('driver_stint_no')
def increment_team_stint_value(key,dict):
    update_value_in_dicts(key, dict, 'team_stint_no', search_key_in_dicts(key, dict).get('team_stint_no') + 1)
    return search_key_in_dicts(key, new_data_pivoted_dict).get('team_stint_no')

key = new_data.at[0, 'key']
update_value_in_dicts(key, new_data_pivoted_dict, 'team_stint_no', 1)
#getting team_stint_no
new_data['team_stint_no'] = 0
for i in range(0,len(new_data)):
    x = new_data.iloc[i]
    if x['change_stint'] == 1 or x['change_team'] == 1:
        increment_team_stint_value(x.key, new_data_pivoted_dict)
    new_data.at[x.name, 'team_stint_no'] = (search_key_in_dicts(x.key, new_data_pivoted_dict).get('team_stint_no'))
```

```python
new_data['driver_stint'] = new_data['driver_name'] + " Stint #" +  new_data['driver_stint_no'].map(int).map(str)
new_data['team_stint'] = new_data['team_no'] + " Stint #" +  new_data['team_stint_no'].map(int).map(str)
#dropping s1_seconds, since the fia wec data seems to add new columns every iteration ;;
new_data = new_data.drop(['change_driver', 'change_team', 'change_stint', 'key', 's1_seconds', 's2_seconds', 's3_seconds'], axis=1)
```
