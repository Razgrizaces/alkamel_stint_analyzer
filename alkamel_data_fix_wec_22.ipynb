```python
import pandas as pd
import numpy as np
alkamel_data = pd.read_csv('alkamel_data_no_duplicates.csv')
```

```python
alkamel_data_1s_lap_time = alkamel_data[alkamel_data['lap_time_seconds']<20]
alkamel_data_1s_lap_time
```

```python
alkamel_data['season'] = alkamel_data['season'].map(str)
alkamel_data_not_wec = alkamel_data[~((alkamel_data['championship']=='FIAWEC')|(alkamel_data['championship']=='FIA WEC'))]
alkamel_data_not_wec['lap_key'] = alkamel_data_not_wec['key']
alkamel_data_wec = alkamel_data[(alkamel_data['championship']=='FIAWEC')|(alkamel_data['championship']=='FIA WEC')]
alkamel_data_wec.groupby('championship').count()
```

```python
alkamel_data_not_wec.groupby('championship').count()
```

```python
alkamel_data_wec['championship'] = 'FIA WEC'
alkamel_data_wec['lap_key'] = alkamel_data_wec['season'] + "_" +alkamel_data_wec['championship'] + "_" + alkamel_data_wec['round'].map(str) +\
"_" + alkamel_data_wec['session'] + "_" + alkamel_data_wec['number'].map(str) + "_" + alkamel_data_wec['class'] + "_" \
    + alkamel_data_wec['driver_number'].map(str) + "_" + alkamel_data_wec['lap_number'].map(str)
alkamel_data_wec
```

```python
full_data_with_wec_fix = pd.concat([alkamel_data_wec, alkamel_data_not_wec])
full_data_with_wec_fix = full_data_with_wec_fix.drop('key', axis=1)
full_data_with_wec_fix
```

```python
full_data_with_wec_fix['season'] = full_data_with_wec_fix['season'].map(str)
full_data_with_wec_fix['round'] = full_data_with_wec_fix['round'].map(str)
full_data_with_wec_fix_race = full_data_with_wec_fix[full_data_with_wec_fix['session_type']=='RACE']
full_data_with_wec_fix_race.groupby(['championship', 'season', 'round']).count().to_csv('check_counts.csv')
```

```python
full_data_with_wec_fix['key'] = full_data_with_wec_fix['lap_key']
full_data_with_wec_fix = full_data_with_wec_fix.drop('lap_key', axis=1)
full_data_with_wec_fix.to_csv('alkamel_data_no_duplicates.csv', index=False)
```

```python
full_data_with_wec_fix_counts = full_data_with_wec_fix.groupby('key').count()
full_data_with_wec_fix_counts[full_data_with_wec_fix_counts['number']>1]
```
