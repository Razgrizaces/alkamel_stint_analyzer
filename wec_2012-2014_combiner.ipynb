```python
import pandas as pd
import numpy as np
```

```python
seasons = ['2012_', '2013_', '2014_', '2015_', '2016_', '2017_', '2018-2019_', '2019-2020_','2021_', '2022_']
season_csv = seasons[8] + "WEC_SEASON.csv"
season = pd.read_csv(season_csv, index_col =0, sep = ";")
season[season['GROUP'].notna()]
```

```python
vehicle_man_lookup = pd.read_csv("vehicle_manufacturer_mapping.csv", index_col=0,sep = ";")
lookup_2014 = pd.read_csv("2014_mapping.csv",index_col=0, sep = ";")
lmp1_lookups = lookup_2014[(lookup_2014['CLASS'] == "LMP1-H") | (lookup_2014['CLASS'] == "LMP1-L")].reset_index(drop=True)
lmp1_lookups
```

```python
other_lookups = lookup_2014[(lookup_2014['CLASS'] != "LMP1-H") & (lookup_2014['CLASS'] != "LMP1-L")].reset_index(drop=True)
wanted_columns = ['VEHICLE', 'MANUFACTURER']
other_lookups = other_lookups[wanted_columns]
other_lookups
```

```python
season_lmp1 = season[season['CLASS'] == "LMP1"].reset_index(drop=True)
season_lmp1
```

```python
season_lmp1_right_class = pd.merge(left = season_lmp1, right = lmp1_lookups, on = "VEHICLE", how = "outer")
season_lmp1_right_class['CLASS'] = season_lmp1_right_class['CLASS_y']
season_lmp1_right_class['MANUFACTURER'] = season_lmp1_right_class['MANUFACTURER_y']
season_lmp1_right_class = season_lmp1_right_class.drop(['CLASS_x', 'MANUFACTURER_x','CLASS_y', 'MANUFACTURER_y'], axis=1)
```

```python
season_not_lmp1 = season[season['CLASS'] != "LMP1"].reset_index(drop=True)
season_not_lmp1
```

```python
season_not_lmp1_right_man = pd.merge(left = season_not_lmp1, right = other_lookups, on = "VEHICLE", how = "outer")
season_not_lmp1_right_man['MANUFACTURER'] = season_not_lmp1_right_man['MANUFACTURER_y']
season_not_lmp1_right_man = season_not_lmp1_right_man.drop(['MANUFACTURER_x','MANUFACTURER_y'], axis=1)
season_not_lmp1_right_man
```

```python
season_combined = season_lmp1_right_class.append(season_not_lmp1_right_man).reset_index(drop="True")
season_combined = season_combined.drop('GROUP', axis=1)
season_combined.to_csv("2014_WEC_SEASON_w_man.csv", sep = ";")
```
