```python
import pandas as pd
import numpy as np
```

```python
#2012=2014 data don't have teams, manufacturer or class

prefix_yr = "2012_"
year = 2012
tracks_2012 = ["SEBRING","SPA_FRANCORCHAMPS","LE_MANS","INTERLAGOS","SILVERSTONE","BAHRAIN_INTERNATIONAL_CIRCUIT","FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT"]
tracks_2013 = ["SILVERSTONE","SPA_FRANCORCHAMPS","LE_MANS","INTERLAGOS","CIRCUIT_OF_THE_AMERICAS", "FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT", "BAHRAIN_INTERNATIONAL_CIRCUIT"]

tracks_2014 = ["SILVERSTONE","SPA_FRANCORCHAMPS","LE_MANS","CIRCUIT_OF_THE_AMERICAS", "FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT", "BAHRAIN_INTERNATIONAL_CIRCUIT","INTERLAGOS"]
tracks_2015 = ["SILVERSTONE","SPA_FRANCORCHAMPS","LE_MANS","NURBURGRING", "CIRCUIT_OF_THE_AMERICAS", "FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT", "BAHRAIN_INTERNATIONAL_CIRCUIT"]
tracks_2016_2017 = ["SILVERSTONE","SPA_FRANCORCHAMPS","LE_MANS","NURBURGRING","AUTODROMO_HERMANOS_RODRIGUEZ", "CIRCUIT_OF_THE_AMERICAS", "FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT", "BAHRAIN_INTERNATIONAL_CIRCUIT"]
tracks_2018_2019 = ["SPA_FRANCORCHAMPS_2018","LE_MANS_2018","SILVERSTONE","FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT","SEBRING","SPA_FRANCORCHAMPS_2019","LE_MANS_2019"]
tracks_2019_2020 = ["SILVERSTONE","FUJI_SPEEDWAY","SHANGHAI_INTERNATIONAL_CIRCUIT","BAHRAIN_INTERNATIONAL_CIRCUIT_2019","CIRCUIT_OF_THE_AMERICAS","SPA_FRANCORCHAMPS","LE_MANS","BAHRAIN_INTERNATIONAL_CIRCUIT_2020"]
tracks_2021 = ["SPA_FRANCORCHAMPS","AUTODROMO_DO_ALGARVE", "AUTODROMO_NAZIONALE_DI_MONZA","LE_MANS","BAHRAIN_INTERNATIONAL_CIRCUIT_6_HOURS", "BAHRAIN_INTERNATIONAL_CIRCUIT_8_HOURS"]
tracks_2022 = ['SEBRING']
wanted_columns = ['NUMBER', 'TEAM', 'VEHICLE', 'CLASS']

classification_df = pd.read_csv("data/classifications/2012_CLASSIFICATION.csv", sep = ";")

classification_df = classification_df[wanted_columns]
print(classification_df)
```

```python
csv = "data/" + prefix_yr + tracks_2012[0] + ".csv"
combined_df = pd.read_csv(csv, index_col = 0)
combined_df.columns =  combined_df.columns.str.strip()
combined_df = combined_df.merge(classification_df, how = "inner")
combined_df.columns = combined_df.columns.str.strip()
combined_df['round'] = 1
combined_df
```

```python
for i in range(1, len(tracks_2012)):
    csv = "data/" + prefix_yr + tracks_2012[i] + ".csv"
    if i >= 6:
        track_df = pd.read_csv(csv, sep=";")
    else:
        track_df = pd.read_csv(csv, index_col=0)
    track_df = track_df.merge(classification_df, how = "inner")
    track_df.columns =  track_df.columns.str.strip()
    track_df['season'] = year
    track_df['round'] = i+1
    track_df['circuit'] = tracks_2012[i]
    combined_df = combined_df.append(track_df)
combined_df
```

```python
combined_df_nas = combined_df[combined_df['TEAM'].isna()]
combined_df_nas
```

```python
combined_df = combined_df.dropna(axis = 1, how = 'all')
combined_df = combined_df.reset_index(drop=True)
combined_df
```

```python
csv_path = prefix_yr + "WEC_SEASON.csv"
combined_df.to_csv(csv_path, sep = ";")
```
