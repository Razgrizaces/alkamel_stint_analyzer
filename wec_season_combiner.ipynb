```python
#for 2015-2022 we have to map to classification to grab vehicle
#seasons = ['2015_', '2016_', '2017_', '2018-2019_', '2019-2020_','2021_', '2022_']
season_path = seasons[0] + "WEC_SEASON.csv"
season = pd.read_csv(season_path, index_col = 0, sep = ";")
wanted_columns = ["NUMBER", 'VEHICLE']
season_classification_path = "data/classifications/" + seasons[0] + "CLASSIFICATION.csv"
season_classification = pd.read_csv(season_classification_path, index_col=0)
season_classification = season_classification[wanted_columns]
season_with_classification = pd.merge(left = season, right = season_classification, on = "NUMBER", how = "outer")
season_with_classification
```

```python
wanted_columns = ["NUMBER", 'VEHICLE']
for i in range(1, len(seasons)):
    season_classification_path = "data/classifications/" + seasons[i] + "CLASSIFICATION.csv"
    season_classification = pd.read_csv(season_classification_path, index_col=0)
    season_classification = season_classification[wanted_columns]
    season_path = seasons[i] + "WEC_SEASON.csv"
    new_season = pd.read_csv(season_path, sep = ";", index_col=0)
    new_season_with_classification = pd.merge(left = new_season, right = season_classification, on = "NUMBER", how = "outer")
    if i !=3 and i !=4:
        new_season_with_classification = new_season_with_classification[new_season_with_classification['season'].notna()]
        new_season_with_classification['season'] = new_season_with_classification['season'].map(int).map(str)
    season_with_classification = season_with_classification.append(new_season_with_classification)
season_with_classification.reset_index(drop=True)
```

```python
full_wec_seasons.columns = full_wec_seasons.columns.str.lower()
full_wec_seasons.columns
```

```python
def convert_to_ms(x):
    hour = 0
    try:
        hour, minutes, second = x.split(":")
    except ValueError:
        minutes, second = x.split(":")
    hour_seconds = int(hour)*60*60
    seconds, ms = second.split(".")
    minute_seconds = (int(minutes)*60 + int(seconds) + int(hour_seconds))*1000 
    total_ms = minute_seconds + int(ms)
    return total_ms

full_wec_seasons['lap_time_ms'] = full_wec_seasons['lap_time'].apply(convert_to_ms)
full_wec_seasons['lap_time_s'] = full_wec_seasons['lap_time_ms']/1000
```
