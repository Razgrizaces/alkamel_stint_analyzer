```python
wanted_columns = ['number', 'class', 'team', 'vehicle', 'manufacturer']
#fix classification on 2012
lemans_2012_class = pd.read_csv('2012_LE_MANS_RACE.csv', index_col=0)
lemans_2012_class.columns = lemans_2012_class.columns.str.strip().str.lower()
lemans_2012_group = lemans_2012_class.groupby(['number', 'team', 'class', 'vehicle']).count().reset_index()
lemans_2012_group['manufacturer'] = lemans_2012_group['vehicle']
lemans_2012_group[wanted_columns].to_csv('lemans_2012_classification.csv')
```

```python
fia_wec_data = pd.read_csv('2012-2022_FIA_WEC_FULL_LAP_DATA.csv', index_col=0)
fia_wec_data['championship'] = 'FIA WEC'
fia_wec_data['session'] = 'RACE'
```

```python
fia_wec_data['team_key'] = fia_wec_data['season'].map(str) + "_" + fia_wec_data['championship'] + "_" + fia_wec_data['round'].map(str)\
     + "_" + fia_wec_data['circuit']+ "_" + fia_wec_data['number'].map(str)
fia_wec_data_grouped = fia_wec_data.groupby(['team_key', 'manufacturer', 'vehicle']).count().reset_index()
wanted_columns = ['team_key', 'manufacturer', 'vehicle']
fia_wec_data_grouped = fia_wec_data_grouped[wanted_columns]
#2020_LeMansCup_4_RACE_98_LMP3_2_52
```

```python
alkamel_data['team_key']= alkamel_data['season'].map(str) + "_" + alkamel_data['championship'] + "_" + alkamel_data['round'].map(str) \
    + "_" + alkamel_data['circuit']+ "_" +alkamel_data['number'].map(str) 
fia_wec_race_alkamel_data = alkamel_data[(alkamel_data['championship']=='FIA WEC')]
```

```python
tk_groupby = pd.merge(fia_wec_data_grouped, fia_wec_race_alkamel_data, on = 'team_key', how = 'outer').groupby('team_key').count()
tk_groupby[(tk_groupby['manufacturer_y']==0)].to_csv('team_keys_groupby.csv')
```
